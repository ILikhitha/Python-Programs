```python
# Pull the "Python " wedge 0.2 from the center of the pie: write a python code for the following graph
import matplotlib.pyplot as plt

import numpy as num 

a=num.array([31,21,23,32])

mylabels=["Python","Maths","English","Chemistry"]

mycolors=["blue","red","green","orange"]

myexplode=[0.2,0,0,0]

plt.pie(a,labels=mylabels,explode=myexplode,colors=mycolors)

plt.show()
```
